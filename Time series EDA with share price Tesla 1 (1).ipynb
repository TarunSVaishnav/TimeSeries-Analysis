# Time Resampling

```python
df_tesla.head()
```

```python
df_tesla.resample(rule='A').min()
```

It will give you all the minimum values for everything and if we use max function than we will get all the maximum value for everything.

```python
#same we can do for max
df_tesla.resample(rule='A').max()
```

Here Rule=A means the year and frequency

```python
#Now lets plot the different values 
#open values
df_tesla.resample(rule='A').max()['Open'].plot()
```

We can see the rise in the open prize from 2019 to 2021

```python
#we can have quaterly data by setting the frequency in as rule = 'QS'
df_tesla.resample(rule='QS').max()['Open'].plot()
```

This Graph shows the quaterly information of the data and we can see some rise and dives here unlike yearly data

```python
#Business end frequency for the data 
df_tesla.resample(rule='BA').max()
```

```python
#lets do some plotting and plot the open prize and see using a bar plot
df_tesla['Open'].resample(rule='QS').mean().plot(kind='bar')
```

```python
df_tesla['High'].rolling(11).max().head(20)
```

Rolling function returns you NAs to the rows you want as here if we see for rolling 11 it os giving NAs till 10th row.

```python
#creating a new column
df_tesla['Open 30 days rolling'] = df_tesla['Open'].rolling(30).mean()
```

```python
df_tesla.head(30)
```

```python
#plotting open with respect to open 30 days rolling
df_tesla[['Open','Open 30 days rolling']].plot(figsize=(12,5))
```

Rolling function can also be used to give the smoothing as it is doing it here 
see the orange line smoothing the graph over blue line.
